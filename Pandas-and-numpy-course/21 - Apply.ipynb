```
import pandas as pd
```

```
bestsellers =  pd.read_csv("/work/bestsellers with categories.csv", sep=",", header=0)
bestsellers.head(5)
```

```
def two_times(value):
    return value * 2
```

```
bestsellers["User Rating"].apply(two_times)
```

```
bestsellers["Rating_2"] = bestsellers["User Rating"].apply(two_times)
bestsellers
```

```
bestsellers["Rating_2"] = bestsellers["User Rating"].apply(lambda x:x * 3)
bestsellers
```

```
bestsellers["Rating_2"] = bestsellers.apply(lambda x : x["User Rating"] * 2 if x["Genre"] == "Fiction" else x["User Rating"], axis=1)
```
