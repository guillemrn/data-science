```
import pandas as pd
import numpy as np
```

```
dict = {
    "Col1": [1,2,3,np.nan],
    "Col2": [4,np.nan,6,7],
    "Col3": ["a","b","c",None],
}
```

```
df = pd.DataFrame(dict)
df
```

```
df.isnull()
```

```
df.isnull()*1
```

```
df.fillna("Missing data")
```

```
# df.fillna(df.mean())
filled_df = df.apply(lambda col: col.fillna(col.mean()) if pd.api.types.is_numeric_dtype(col) else col.fillna("Missing data"), axis=0)
filled_df
```

```
df.interpolate()
```

```
df.dropna()
```
