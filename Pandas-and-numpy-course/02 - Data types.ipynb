```
import numpy as np
```

```
arr = np.array([1,2,3,4])
arr.dtype
```

```
arr = np.array([1,2,3,4], dtype="float64")
arr.dtype

```

```
arr
```

```
arr.astype(np.float64)
```

```
arr = np.array([0,1,2,3,4])
arr.astype(np.bool_)
```

```
arr = np.array([0,1,2,3,4])
arr.astype(np.string_)
```
