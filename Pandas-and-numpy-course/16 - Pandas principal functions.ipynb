```
import pandas as pd
```

```
bestsellers =  pd.read_csv("/work/bestsellers with categories.csv", sep=",", header=0)
bestsellers.head(5)
```

```
# info() nos muestra un resumen de nuestro dataset, incluyendo datos como n√∫mero de columnas, si tiene valores nulos, etc.
bestsellers.info()
```

```
# describe() nos da informacion sobre los datos numericos de nuestro dataset, como mediana, desviacion std, etc.
bestsellers.describe()
```

```
# Nos mostrara los ultimos valores de nuestro dataset, podemos pasarle el numero de elementos como parametro
bestsellers.tail(5)
```

```
# Funciona de la misma forma que tail() pero con los primeros elementos
bestsellers.head(5)
```

```
# La funci√≥n memory_usage() se utiliza para calcular el uso de memoria de cada columna en el DataFrame bestsellers,
# y el par√°metro deep=True indica que se debe incluir el uso de memoria de objetos referenciados en las columnas,
# como por ejemplo, objetos almacenados en celdas de tipo object.

bestsellers.memory_usage(deep=True)
```

```
# value_counts() nos contara el numero de veces que un elemento exista
bestsellers["Author"].value_counts()
```

```
bestsellers.iloc[0]
```

```
bestsellers = bestsellers.append(bestsellers.iloc[0])
```

```
bestsellers
```

```
# drop_duplicates() nos eliminara si existen elementos repetidos en nuestro dataset, podemos pasarle el parametro
# keep="last" para que mantenga el ultimo elementos duplicado y borre los demas.
bestsellers.drop_duplicates()
```

```
# sort_values() ordena el DataFrame por la columna 'Year' en orden descendente (mayor a menor)
bestsellers.sort_values("Year", ascending=False)
```
